# BMI 身體質量指數計算器 - 考題

## 題目說明

請設計一個 BMI（Body Mass Index，身體質量指數）計算程式，程式需要：

1. **輸入資料**：
   - 請使用者輸入身高（單位：公分，cm）
   - 請使用者輸入體重（單位：公斤，kg）

2. **計算 BMI**：
   - BMI 計算公式：`BMI = 體重(kg) / 身高(m)²`
   - 注意：輸入的身高是公分，計算時需要轉換為公尺（1 公尺 = 100 公分）

3. **判斷健康狀態**：
   根據計算出的 BMI 值，判斷並顯示健康狀態：
   - BMI < 18.5：體重過輕
   - 18.5 ≤ BMI < 24：正常範圍
   - 24 ≤ BMI < 27：體重過重
   - BMI ≥ 27：肥胖

4. **輸出結果**：
   - 顯示計算出的 BMI 值（保留小數點後 2 位）
   - 顯示對應的健康狀態
